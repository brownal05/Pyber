```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.cm as cm

# File to Load (Remember to change these)
city_data_to_load = ("data/city_data.csv")
ride_data_to_load =("data/ride_data.csv")

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load) 
ride_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
city_df = city_df.drop_duplicates("city", keep = "first")
merged_df = city_df.merge(ride_df,  on = 'city')


# Display the data table for preview
```

```python
merged_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#The ride sharing bonanza continues! Seeing the success of notable players like Uber and Lyft, you've decided to join a fledgling ride sharing company of your own. 
#In your latest capacity, you'll be acting as Chief Data Strategist for the company. In this role, you'll be expected to offer data-backed guidance on new opportunities for market differentiation.

#You've since been given access to the company's complete recordset of rides. This contains information about every active driver and historic ride, including details like city, driver count, individual fares, and city type.

#Your objective is to build a [Bubble Plot](https://en.wikipedia.org/wiki/Bubble_chart) that showcases the relationship between four key variables:

#* Average Fare ($) Per City
#* Total Number of Rides Per City
#* Total Number of Drivers Per City
#* City Type (Urban, Suburban, Rural)

city = merged_df.groupby("city") 
avg_fare = city.mean()["fare"]
ride_count = city.count()["ride_id"]
driver_count = city.mean()["driver_count"]
city_type = city_df.set_index("city")["type"]

cities = pd.DataFrame({"Average Fare": avg_fare,
                       "Number of Rides": ride_count,
                       "Number of Drivers": driver_count,
                       "City Type": city_type})
#print(city_type)
urban = cities[cities['City Type'] == 'Urban']
suburban = cities[cities['City Type'] == 'Suburban']
rural = cities[cities['City Type'] == 'Rural']
# Build the scatter plots for each city types

color_scheme = {"Gold": "#FFDF00", "Light Sky Blue": "#87CEFA", "Light Coral": "#F08080"}
city_color = {"Rural" : color_scheme["Gold"], "Suburban": color_scheme["Light Sky Blue"], "Urban": color_scheme["Light Coral"]}
plt.scatter(urban["Number of Rides"], urban["Average Fare"],  color = city_color["Urban"], s = urban["Number of Drivers"]*10, alpha=0.6, label = "Urban", linewidth = 1)
plt.scatter(suburban["Number of Rides"], suburban["Average Fare"],  color = city_color["Suburban"], s = suburban["Number of Drivers"]*10, alpha=0.6, label = "Suburban", linewidth = 2)
plt.scatter(rural["Number of Rides"], rural["Average Fare"],  color = city_color["Rural"], s = rural["Number of Drivers"]*10, alpha=0.6, label = "Rural", linewidth = 4)
plt.suptitle('Pyber Ride Data')

plt.grid()
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.legend(loc = "best", fontsize = "small", scatterpoints = 1,  markerscale= .5, scatteryoffsets=[0.5, 0.5, 0.75], fancybox = True)
plt.suptitle("Pyber Ride Sharing Data (2016)")
plt.title("(Dot size indicates number of drivers)")
plt.savefig("Images/Scatter.png")

plt.show()

                
# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

## Total Fares by City Type

```python
# Calculate Type Percents
cityType = merged_df.groupby("type")["type", "fare", "ride_id", "driver_count"]
fare_avg = cityType.sum()["fare"]
# Build Pie Chart


 
plt.pie(fare_avg, startangle = 145, colors = ["#FFDF00", "#87CEFA", "#F08080"], explode = [0 , 0, .1], labels = fare_avg.index, autopct="%1.1f%%", shadow=True)
plt.title('% of Total Fares by City Type')

# Save Figure
plt.savefig("Images/fare.png")

plt.show()
```
